```python
from glob import glob
import pandas as pd

#files = glob("./edailynewsprocessed//*.csv")
files = glob("./edailynewsprocessed/*.csv")
base_df = pd.read_csv("rate.csv")

filenum = 0

    
col_names =  ['date', 'ngram+unigram']
my_df  = pd.DataFrame(columns = col_names)

for file in files:
    df = pd.read_csv(file,lineterminator='\n')
    
    aa = df['token']
    bb = df['ngram']
    cc = df['date']

    
    for i, row in df.iterrows() :
        try :
            
            tokenngram = str(aa[i])+","+str(bb[i])
            
            #print(tokenngram)
            my_df.loc[filenum*10+i, 'date'] = str(cc[i])
            my_df.loc[filenum*10+i, 'ngram+unigram'] = tokenngram

            print("success")
            if i % 10 == 0 :
                print('{}/{} : 성공'.format(i, df.shape[0]))
        except Exception as ex :
            print("fail")
            #print("{} : {}".format(i, str(ex)))

    filenum += 1
    print(filenum)

    #my_df.to_csv("NBC.csv", encoding='UTF-8', mode='w')

new_df = pd.merge(base_df, my_df, on='date')


final_df = pd.merge(base_df, my_df, how='inner', on='date')
final_df = final_df.drop(columns=['rate','difference'])
final_df = final_df[['date', 'ngram+unigram', 'label']]
print(final_df)
#final_df.to_csv("NBC.csv", encoding='UTF-8', mode='w')
```

```python
import pandas as pd
from glob import glob
import datetime
```

```python
call = pd.read_csv('rate.csv', encoding='utf-8', names=['date', 'rate', 'difference',' label'], dtype={'date':str})
tmp = [] 
ls = call['date']
for i in ls:
    i = str(i).replace(' ', '')
    tmp.append(i)
call['date'] = tmp

date_ls = call['date']
date_c = []
for i in date_ls :
    i = str(i).replace('.','-')
    date_c.append(i)
call['date']=date_c
```

```python
file_ls = glob('./edailynewsprocessed/*.csv')
files = []
for i in file_ls:
    i= i.replace('\\','/')
    files.append(i)

    
```

```python
df = pd.DataFrame()
for file_name in files:
    file = pd.read_csv(file_name, encoding = 'utf-8', names=['com','date', 'title','url','script','token','ngram'])
    tmp = [] 
    ls = file['date']
    for i in ls:
        i = str(i).replace(' ', '')
        tmp.append(i)
    file['date'] = tmp
    df = df.append(file)
    print('========================================================')
    print(len(df))
```

```python
file = df
a = file['token']
b = file['ngram']
file['token+ngram'] = a+b
```

```python
df_new = pd.merge(file,call, left_on="date", right_on="date")
df_new
```
